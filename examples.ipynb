````python
import matplotlib.pyplot as plt
import numpy as np

def ricker(x, r):
    return x * np.exp(r * (1 - x))

r = np.linspace(0, 4, 1000)

x = 1e-5 * np.ones(1000)

length = 1000

for i in range(length):
    x = ricker(x, r)
    if i >= (length - 100):
        plt.plot(r, x, ',k', alpha=.25)

plt.xlim(0, 4)
plt.title("Bifurcation diagram of the Ricker model")
plt.xlabel('r')
plt.ylabel('Population')
plt.show()
````
